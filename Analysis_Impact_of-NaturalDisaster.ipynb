```python
import statsmodels.api as sm

#Correlation Analysis
natrualD = pd.read_csv('data_1900-2021.csv')
GDP = pd.read_csv('GDP.csv')
# Convert relevant columns to numeric values
naturalD['Total Affected'] = pd.to_numeric(naturalD['Total Affected'], errors='coerce')
naturalD["Total Damages ('000 US$)"] = pd.to_numeric(naturalD["Total Damages ('000 US$)"], errors='coerce')
naturalD['Total Deaths'] = pd.to_numeric(naturalD['Total Deaths'], errors='coerce')

# Create a flag column for missing values
naturalD['Missing_Data_Flag'] = naturalD[['Total Affected', 'Total Deaths', "Total Damages ('000 US$)"]].isnull().any(axis=1).astype(int)

# Function to fill missing values with country-disaster-type median, then disaster-type median
def fill_missing_values(column):
    naturalD[column] = naturalD.groupby(['Country', 'Disaster Type'])[column].transform(lambda x: x.fillna(x.median()))
    return naturalD[column]

# Apply the function to relevant columns
naturalD['Total Affected'] = fill_missing_values('Total Affected')
naturalD['Total Deaths'] = fill_missing_values('Total Deaths')


# Assume zero if missing because is safer than using an inflated estimate
naturalD["Total Damages ('000 US$)"] = naturalD["Total Damages ('000 US$)"].fillna(0)

# Compute the disaster severity index
naturalD['Disaster_Severity'] = (naturalD['Total Affected'] +
                           10 * naturalD['Total Deaths'] +
                          naturalD["Total Damages ('000 US$)"] / 1_000_000)

# Aggregate disaster severity by country & year (sum of all disasters)
df_severity = naturalD.groupby(['Country', 'Year'])['Disaster_Severity'].sum().reset_index()
Data_All = pd.merge(GDP, naturalD, on=['Year', 'Country'])
# Create a lagged disaster severity variable (previous year's severity)
Data_All['Lagged_Disaster_Severity'] = Data_All.groupby('Country')['Disaster_Severity'].shift(1)
# Calculate GDP growth (GDP)
Data_All['GDP_Growth'] = Data_All.groupby('Country')['Value'].pct_change() * 100  # Percentage change
# Drop rows with missing values (due to lagging and growth calculations)
Data_All = Data_All.dropna(subset=['GDP_Growth', 'Lagged_Disaster_Severity'])

# Regression analysis: Does past disaster severity affect GDP growth?
X = Data_All[['Lagged_Disaster_Severity']]  # Independent variable
X = sm.add_constant(X)  # Add intercept
y = Data_All['GDP_Growth']  # Dependent variable

# Fit the regression model
model = sm.OLS(y, X).fit()

# Output the results
print(model.summary())
```
